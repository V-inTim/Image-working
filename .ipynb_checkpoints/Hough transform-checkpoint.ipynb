```python
import numpy as np
import cv2
from IPython.display import Image
import math
img = cv2.imread('cube.png')
```

```python
filt = cv2.bilateralFilter(img,9,75,75)
edges = cv2.Canny(filt,50,200,apertureSize = 3)
lines = cv2.HoughLines(edges,1, np.pi / 180, 70, None, 0, 0)

print(lines)
```

```python
for i in range(len(lines[:,0,0])):
    for rho,theta in lines[i]:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 500*(-b))
        y1 = int(y0 + 500*(a))
        x2 = int(x0 - 500*(-b))
        y2 = int(y0 - 500*(a))
        cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)
cv2.imwrite('hough.jpg',img)
Image('hough.jpg', width=320, height=240)
```
